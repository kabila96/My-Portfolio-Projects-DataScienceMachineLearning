### Objective:

This project involves conducting People Analytics and HR Analysis for a multimillion-dollar business. The analysis aims to understand employee performance, turnover, satisfaction, and other HR metrics to make informed decisions for business success.




The dataset will include the following columns:

Satisfaction_Level: Job satisfaction rating (1-10 scale).
                                             
Training_Hours: Number of hours spent on training in the past year.

Job_Role: Employee's job role (e.g., Software Engineer, Sales Executive, HR Specialist).

Travel_Frequency: Frequency of business travel (None, Rare, Frequent).

# 1.Predictive Modeling: Predicting Attrition

We will now apply a Random Forest Classifier to predict employee attrition based on available features. 

This will help us understand the factors influencing attrition

### Data Preprocessing & Feature Selection:

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Encoding categorical features
df_encoded = pd.get_dummies(df, drop_first=True).astype(int)
```

```python
df_encoded.head()
```

```python
df_encoded.columns
```

```python
# Splitting data into training and testing sets
X = df_encoded.drop("Attrition", axis=1)
y = df_encoded["Attrition"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Random Forest Classifier Model
rf_model = RandomForestClassifier(n_estimators=500, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# Predicting on test data
y_pred = rf_model.predict(X_test)
```

```python
# Evaluate model performance
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix:\n{conf_matrix}")
```

```python
# Classification Report
print(f"Classification Report:\n{classification_report(y_test, y_pred)}")
```

# Feature Importance Analysis

#### We will now analyze the importance of different features in predicting employee attrition.

```python
# Feature Importance from the Random Forest model
feature_importances = rf_model.feature_importances_
```

```python
# Create a DataFrame to view the feature importance
importance_df = pd.DataFrame({
    "Feature": X.columns,
    "Importance": feature_importances
}).sort_values(by="Importance", ascending=False)
```

```python
importance_df
```

```python
# Plot the feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x="Importance", y="Feature", data=importance_df, palette="viridis")
plt.title("Feature Importance for Predicting Attrition")
plt.show()
```
