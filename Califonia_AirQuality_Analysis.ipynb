## GRAPHING AIR QUALITY INDEX(AQI)data

## Plot AQI vs Time

```python
plt.figure(figsize=(12,5))
sns.lineplot(x='Date',y='daily_aqi_value', data=df,marker='o')
plt.title('Air Quality Index Over Time')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show();
```

```python
plt.figure(figsize=(12,5))
sns.lineplot(x='daily_mean_pm2_5_concentration',y='daily_aqi_value', data=df,marker='o')
plt.title('AQI vs PM 2.5 Concetration')
plt.xlabel('Daily Mean PM2.5 Concentration')
plt.ylabel('Daily AQI Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show();
```

## Linear Regression to Evaluate AQI vs PM2.5 Correlation

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

## prepare data for linear regression

```python
X = df[['daily_mean_pm2_5_concentration']].values
y = df['daily_aqi_value'].values
```

### create and fit the model

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

### Predict AQI Values

```python
y_pred = model.predict(X)
```

```python
y_pred
```

### calculate r2 score

```python
r2 = r2_score(y, y_pred)
print(f'R^2 value: {r2:4f}')
```

## 85%

```python
plt.figure(figsize=(10,6))

#Scatter plot
sns.scatterplot(x='daily_mean_pm2_5_concentration',y='daily_aqi_value',data=df, label='Data')

#RegressionLine
sns.lineplot(x=df['daily_mean_pm2_5_concentration'],y=y_pred,color='red', label='Regression Line')

plt.title(f'AQI vs PM2.5 (R^2={r2:.4f}')
plt.xlabel('PM2.5')
plt.ylabel('AQI')
plt.legend()
plt.show();
```

### Extract Day and Month

```python
df_2000['Month_Day'] = df_2000['Date'].dt.strftime('%m-%d')
df_2024['Month_Day'] = df_2024['Date'].dt.strftime('%m-%d')
```

```python
df_2000.head()
```

```python
plt.figure(figsize=(12,8))

# Plot year 1 (2024)
sns.scatterplot(x='Month_Day', y='daily_aqi_value', data=df_2024, 
                color='blue', label='2024', alpha=0.6)  

# Plot year 2 (2000)
sns.scatterplot(x='Month_Day', y='daily_aqi_value', data=df_2000, 
                color='red', label='2000', alpha=0.6)

plt.xticks(rotation=45)
plt.title('Daily AQI Value vs Month-Day for Two different Years (2024 and 2000)')
plt.ylabel('Daily AQI Value')
plt.legend()
plt.show()
```

```python
df_2024['Month'] = df_2024['Date'].dt.month
df_2024['Year'] = df_2024['Date'].dt.year

df_2000['Month'] = df_2000['Date'].dt.month
df_2000['Year'] = df_2000['Date'].dt.year
```

```python
df_2000.head(2)
```

##### THE GRAPH ABOVE LOOKS VERY BUSSY HENCE WE NEED TO SEPARATE THESE BY CALCULATING THE AVERAGE MEAN FOR EACH YEARS 2024 AND 2000

```python
df_2000.columns
```

```python
#calculate monthly average AQI for year 1 (2024)
monthly_avg_2024 = df_2024.groupby(['Year', 'Month'])['daily_aqi_value'].mean().reset_index()

#calculate monthly average AQI for year 2 (2000)
monthly_avg_2000 = df_2000.groupby(['Year', 'Month'])['daily_aqi_value'].mean().reset_index()
```

## BUILD A LINE PLOT SHOWING THE TWO  YEARS EMMISSIONS

```python
plt.figure(figsize=(12, 8))

#plot for year 1 (2024)
sns.lineplot(x='Month', y='daily_aqi_value', data=monthly_avg_2024, marker='o', color='blue', label='2024')

#plot for year 2 (2000)
sns.lineplot(x='Month', y='daily_aqi_value', data=monthly_avg_2000, marker='o', color='red', label='2000')

plt.title('Monthly Average AQI vs Month for Two Different Years')
plt.xlabel('Month')
plt.ylabel('Monthly Average AQI')
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.show()
```
